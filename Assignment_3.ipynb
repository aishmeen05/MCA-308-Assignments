```python
import pandas as pd
from scipy.stats import ttest_ind

# Load the data
df = pd.read_csv('AWSales.csv')

# Separate buyers vs non-buyers using the numeric column
buyers_spend = df[df['BikeBuyer'] == 1]['AvgMonthSpend']
nonbuyers_spend = df[df['BikeBuyer'] == 0]['AvgMonthSpend']

# Compute averages
print("Average monthly spend of buyers:", buyers_spend.mean())
print("Average monthly spend of non-buyers:", nonbuyers_spend.mean())

# Two-sample t-test
t_stat, p_value = ttest_ind(buyers_spend, nonbuyers_spend, equal_var=False)  # Welch's t-test
print("t-statistic:", t_stat)
print("p-value:", p_value)

# Interpretation
if p_value < 0.05:
    print("Significant difference in average monthly spend between buyers and non-buyers.")
else:
    print("No significant difference in average monthly spend between buyers and non-buyers.")
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load data
df = pd.read_csv('AWSales.csv')

# Ensure BikeBuyer is numeric
df['BikeBuyer'] = df['BikeBuyer'].astype(int)

# Set plot style
sns.set(style="whitegrid")

# 1️⃣ Boxplot
plt.figure(figsize=(8,6))
sns.boxplot(x='BikeBuyer', y='AvgMonthSpend', data=df)
plt.xticks([0,1], ['Non-Buyer', 'Buyer'])
plt.title('Monthly Spend of Bike Buyers vs Non-Buyers (Boxplot)')
plt.ylabel('Average Monthly Spend')
plt.xlabel('Customer Type')
plt.show()

# 2️⃣ Violin plot
plt.figure(figsize=(8,6))
sns.violinplot(x='BikeBuyer', y='AvgMonthSpend', data=df)
plt.xticks([0,1], ['Non-Buyer', 'Buyer'])
plt.title('Monthly Spend Distribution (Violin Plot)')
plt.ylabel('Average Monthly Spend')
plt.xlabel('Customer Type')
plt.show()

# 3️⃣ Bar plot with mean and standard deviation
mean_spend = df.groupby('BikeBuyer')['AvgMonthSpend'].mean()
std_spend = df.groupby('BikeBuyer')['AvgMonthSpend'].std()

plt.figure(figsize=(6,5))
plt.bar(['Non-Buyer', 'Buyer'], mean_spend, yerr=std_spend, capsize=5, color=['lightcoral','lightblue'])
plt.title('Average Monthly Spend with Standard Deviation')
plt.ylabel('Average Monthly Spend')
plt.show()
```
