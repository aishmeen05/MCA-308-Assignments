```python
import pandas as pd

# Create DataFrame from your data
data = {
    'Employee_ID': [101, 102, 103, 104, 105],
    'Name': ['Alice', 'Bob', 'Charlie', 'Diana', 'Edward'],
    'Department': ['HR', 'IT', 'IT', 'Marketing', 'Sales'],
    'Age': [29, 34, 41, 28, 38],
    'Salary': [50000, 70000, 65000, 55000, 60000],
    'Years_of_Experience': [4, 8, 10, 3, 12],
    'Joining_Date': ['2020-03-15', '2017-07-19', '2013-06-01', '2021-02-10', '2010-11-25'],
    'Gender': ['Female', 'Male', 'Male', 'Female', 'Male'],
    'Bonus': [5000, 7000, 6000, 4500, 5000],
    'Rating': [4.5, 4.0, 3.8, 4.7, 3.5]
}

df = pd.DataFrame(data)

# a) Shape of DataFrame
print("Shape:", df.shape)

# b) Summary of DataFrame
print("\nInfo:")
print(df.info())

# c) Descriptive statistics
print("\nDescriptive statistics:")
print(df.describe())

# d) First 5 rows and last 3 rows
print("\nFirst 5 rows:\n", df.head())
print("\nLast 3 rows:\n", df.tail(3))

# e) Statistics
average_salary = df['Salary'].mean()
total_bonus = df['Bonus'].sum()
youngest_age = df['Age'].min()
highest_rating = df['Rating'].max()

print("\nAverage Salary:", average_salary)
print("Total Bonus:", total_bonus)
print("Youngest Age:", youngest_age)
print("Highest Rating:", highest_rating)

# f) Sort by Salary descending
df_sorted = df.sort_values(by='Salary', ascending=False)
print("\nSorted by Salary (desc):\n", df_sorted)

# g) Categorize employees by Rating
def categorize_rating(r):
    if r >= 4.5:
        return 'Excellent'
    elif r >= 4.0:
        return 'Good'
    else:
        return 'Average'

df['Performance_Category'] = df['Rating'].apply(categorize_rating)
print("\nWith Performance_Category:\n", df)

# h) Missing values
print("\nMissing values:\n", df.isnull().sum())

# i) Rename Employee_ID to ID
df.rename(columns={'Employee_ID': 'ID'}, inplace=True)
print("\nRenamed column:\n", df.head())

# j) Find employees
# i. More than 5 years experience
exp_employees = df[df['Years_of_Experience'] > 5]
print("\nEmployees with >5 years experience:\n", exp_employees)

# ii. IT department
it_employees = df[df['Department'] == 'IT']
print("\nEmployees in IT:\n", it_employees)

# k) Add Tax column (10% of Salary)
df['Tax'] = df['Salary'] * 0.10
print("\nWith Tax column:\n", df)

# l) Save modified DataFrame to CSV
df.to_csv('employees_modified.csv', index=False)
print("\nModified DataFrame saved as 'employees_modified.csv'")
```
